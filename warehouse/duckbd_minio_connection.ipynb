```python
import duckdb 
```

```python
conn = duckdb.connect("warehouse.duckdb") # apre una connessione a un database DuckDB chiamato "warehouse.duckdb"
conn.execute("INSTALL httpfs;") # installa il modulo httpfs per consentire a DuckDB di leggere dati da fonti esterne come S3
conn.execute("LOAD httpfs;") # carica il modulo httpfs per consentire a DuckDB di leggere dati da fonti esterne come S3

# Configurazione delle credenziali e dell'endpoint S3 per DuckDB
conn.execute("""
    SET s3_endpoint='localhost:9000';
    SET s3_access_key_id='minioadmin';
    SET s3_secret_access_key='minioadmin';
    SET s3_use_ssl=false;
    SET s3_url_style='path';
""") # 

result = conn.execute("SELECT * FROM read_parquet('s3://staging/prices/AAPL_prices.parquet') LIMIT 5").fetchdf() # esegue una query per leggere i dati dal file Parquet su S3 e li restituisce come DataFrame
print(result)
```

```python
conn.close() # chiude la connessione al database DuckDB
```
